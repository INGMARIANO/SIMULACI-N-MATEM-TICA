1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función

$$ f(x)=x^2-4x+2 $$

```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
x=np.linspace(-10,10,1000)
y=x**2-4*x+2
plt.plot(x,y)
plt.grid()

```

```python
sp.var("x")
sp.var("y")
y=x**2-4*x+2
pc=sp.solve(y,x)
print("Los valores donde y=0 son:", pc)
print("Al ser cuadratica, su dominio está dentro de todos los números reales")
```

$$ f(x)=\sin(3x) $$

```python
sp.var("x")
def f(x):
    return np.sin(3*x)
x=np.linspace(-0,2*np.pi,1000)
plt.plot(x,f(x))
plt.grid()
```

```python
sp.var("x")
sp.var("y")
y=sp.sin(3*x)
pc=sp.solve(y,x)
print("Los valores donde y=0 son:", pc)
print("Al ser una función de seno, su dominio está dentro de todos los números reales")
```

$$ f(x)= e^2x-3 $$

```python
sp.var("x")
def g(x):
    return np.e**(2*x)-3
x=np.linspace(-50,50,1000)
plt.plot(x,g(x))
plt.grid()
```

```python
sp.var("x")
sp.var("y")
y=sp.exp(2*x)-3
pc=sp.solve(y,x)
print("Los valores donde y=0 son:", pc)
print("Al ser una función exponencial, su dominio está en todos los números reales")
```

$$ f(x)= \sqrt (x^2-4) $$

```python
sp.var("x")
def h(x):
    return np.sqrt(x**2-4)
x=np.linspace(-50,50,100)
plt.plot(x,h(x))
plt.grid()
```

```python
sp.var("x")
sp.var("y")
y=(x**2-4)**0.5
pc=sp.solve(y,x)
print("Su dominio está dentro de los números reales, con excepción del intervalo", pc)
```

$$ f(x)=\cos(x+\pi/4) $$

```python
sp.var("x")
def j(x):
    return np.cos(x+(np.pi/4))
x=np.linspace(0,2*np.pi,1000)
plt.plot(x,j(x))
plt.grid()
```

```python
sp.var("x")
sp.var("y")
y=sp.cos(x+(sp.pi/4))
pc=sp.solve(y,x)
print("Al ser una función de coseno, su dominio está dentro de todos lo números reales")
print("Sus valores donde y=0, son ", pc)
```

2. Graficar los siguientes grupos de funciones:

Grupo 1
$$ y=2x+1 $$
$$ y=-x+4 $$
$$ y=x^2 $$

```python
def f(x):
    return 2*x+1
def g(x):
    return -x+4
def i(x):
    return x**2
x=np.linspace(-50,50,100)
plt.plot(x,f(x),label="2x+1")
plt.plot(x,g(x), label="-x+4")
plt.plot(x,i(x), label="x^2")
plt.grid()
plt.legend(loc="best")
```

Grupo 2
$$ y= \sin(x) $$
$$ y= \cos(x) $$

```python
def f(x):
    return np.sin(x)
def g(x):
    return np.cos(x)
x=np.linspace(0,2*np.pi,1000)
plt.plot(x,f(x), label="sin(x)")
plt.plot(x,g(x), label="cos(x)")
plt.grid()
plt.legend(loc="best")
```

3. Crea una función que cree una gráfica y dibuje "k" líneas de forma $$ f(x)=\sin (ax) $$

La función recibe:
* x1: Valor inicial de "x"
* x2: Valor final de "x"
* n: cantidad de puntos del vector "x"
* a: Valor inicial de a
* b: Valor final de a
* k: Cantidad de líneas
Usar la variable opcional de "label2 y el comando de legend para saber qué línea pertenece a qué a.

```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
sp.var("x")
def f(x1,x2,n,a,b,k):
    x=np.linspace(x1,x2,n)
    for k in range(0,k):
        plt.plot(x,np.sin(a*x),label=a)
        a-=1
    plt.grid()
    plt.legend(loc="best")
    
    
```
